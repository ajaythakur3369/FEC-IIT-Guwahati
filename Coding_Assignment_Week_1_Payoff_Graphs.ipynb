# **Payoff from Call and Put Options**

# **Call Option**
payoff = max(-C, S - X - C)


# **Put Option**
payoff = max(-P, X - S - P)

P and C are Premiums of Calls and Puts

S is the Stock Price

X is the Strike Price

```python
def payoff(option_type, premium, S, X):
    if option_type == 'C':
        return max(-premium, S - X - premium)
    elif option_type == 'P':
        return max(-premium, X - S - premium)
    else:
        return 'option type not supported'
```

Checking the Output of Payoff Function

```python
# Answer displayed should be 10
payoff('C', 10, 120, 100)
```

```python
# Answer displayed should be option type not supported
payoff('Z', 10, 120, 100)
```

```python
# Answer displayed should be -10
payoff('P', 10, 120, 100)
```

# **Payoff Graphs**

**Call Option**

```python
X = dfo_exp.calls['strike'][0]
premium = dfo_exp.calls['lastPrice'][0]
```

```python
StockPrice = np.arange(0, 6000)
Payoff = [payoff('C', premium, i, X) for i in StockPrice]
```

```python
# Plotting the Payoff Graph for Call Option
plt.rcParams['figure.figsize'] = (15, 6)
ax = plt.axes()
ax.set_facecolor("aliceblue")
plt.plot(StockPrice, Payoff, label = 'Payoff Line')
plt.grid(True, color = "grey", linewidth = "0.4")
plt.xlabel('Stock Price', fontname = 'serif', fontsize = 14)
plt.xticks(fontname = 'serif')
plt.ylabel('Payoff', fontname = 'serif', fontsize = 14)
plt.yticks(fontname = 'serif')
plt.title('Call Option Payoff at Expiry', fontname = 'serif', fontsize = 16)

# Breakeven point is when payoff is 0 or S = X+C
plt.plot(X + premium, 0, 'ro', label = 'Breakeven Point')
plt.legend()
plt.show()
```

It calculates the Data and plots the Payoff Graphs for Call Options.

**Put Option**

```python
X = dfo_exp.puts['strike'][len(dfo_exp.puts) - 1]
premium = dfo_exp.puts['lastPrice'][len(dfo_exp.puts) - 1]
```

```python
StockPrice = np.arange(0, 6000)
Payoff = [payoff('P', premium, i, X) for i in StockPrice]
```

```python
# Plotting the Payoff Graph for put option
plt.rcParams['figure.figsize'] = (15, 6)
ax = plt.axes()
ax.set_facecolor("aliceblue")
plt.plot(StockPrice, Payoff, label = 'Payoff Line')
plt.grid(True, color = "grey", linewidth = "0.4")
plt.xlabel('Stock Price', fontname = 'serif', fontsize = 14)
plt.xticks(fontname = 'serif')
plt.ylabel('Payoff', fontname = 'serif', fontsize = 14)
plt.yticks(fontname = 'serif')
plt.title('Put Option Payoff at Expiry', fontname = 'serif', fontsize = 16)

# Breakeven Point is when Payoff is 0 or S = X - P
plt.plot(X - premium, 0, 'ro', label = 'Breakeven Point')
plt.legend()
plt.show()
```

It calculates the Data and plots the Payoff Graphs for the Put Option.
